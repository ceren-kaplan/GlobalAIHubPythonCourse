```python
puan=0
soru=input("1.Türkiye'nin başkenti neresidir?")
if(soru.lower()=="ankara"):
    print("Doğru cevap verdiniz.")
    puan+=10
else:
    print("Yanlış cevap verdiniz, doğru cevap: ANKARA")
    
soru2=input("2.Duvara asılı bir haritanın sağı her zaman hangi yönü gösterir ?")    
if(soru2.lower()=="doğu"):
    print("Doğru cevap verdiniz.")
    puan+=10
else:
    print("Yanlış cevap verdiniz, doğru cevap: DOĞU")  
    
soru3=input("3.Kızınca tüküren hayvan hangisidir?")    
if(soru3.lower()=="lama"):
    print("Doğru cevap verdiniz.")
    puan+=10
else:
    print("Yanlış cevap verdiniz, doğru cevap: LAMA")
    
soru4=input("4.Gülü ile meşhur olan ilimiz hangisidir?")
if(soru4.replace("I", "ı").replace("İ", "i").lower()=="ısparta"):
    print("Doğru cevap verdiniz.")
    puan+=10
else:
    print("Yanlış cevap verdiniz, doğru cevap: ISPARTA")
    
soru5=input("5.Osmanlı Devletinin kurucusu olan Osmanlı ailesi hangi Türk boyuna mensuptur?")
if(soru5.replace("I", "ı").replace("İ", "i").lower()=="kayı"):
    print("Doğru cevap verdiniz.")
    puan+=10
else:
    print("Yanlış cevap verdiniz, doğru cevap: KAYI")
    
soru6=input("6.Türkiye'de ağaçlandırma ve erozyonla mücadele etmek için kurulan Sivil Toplum Kuruluşu hangisidir?")   
if(soru6.lower()=="tema"):
    print("Doğru cevap verdiniz.")
    puan+=10
else:
    print("Yanlış cevap verdiniz, doğru cevap: TEMA")
    
soru7=input("7.Kuvvet birimi nedir?")
if(soru7.lower()=="newton"):
    print("Doğru cevap verdiniz.")
    puan+=10
else:
    print("Yanlış cevap verdiniz, doğru cevap: NEWTON")
    
soru8=input("8.Yeni bir işe girerken, iş yerlerine verdiğimiz kısa öz geçmişin kısa adı nedir?")   
if(soru8.lower()=="cv"):
    print("Doğru cevap verdiniz.")
    puan+=10
else:
    print("Yanlış cevap verdiniz, doğru cevap: CV")
    
soru9=input("9.Osmanlı hukukunda, mahkemelerde yargılamayı yapan görevlinin adı nedir?")
if(soru9.replace("I", "ı").replace("İ", "i").lower()=="kadı"):
    print("Doğru cevap verdiniz.")
    puan+=10
else:
    print("Yanlış cevap verdiniz, doğru cevap: KADI")
    
soru10=input("10.İlk osmanlı para birimi nedir?")
if(soru10.lower()=="akçe"):
    print("Doğru cevap verdiniz.")
    puan+=10
else:
    print("Yanlış cevap verdiniz, doğru cevap: AKÇE")
    
if(puan>50):
    print("Tebrikler Başardınız,Puanınız:",puan)
else: 
    print("Başarısız,puanınız:",puan)    

```
